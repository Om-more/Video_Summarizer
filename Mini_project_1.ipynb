```python
from yt_dlp import YoutubeDL
import requests

def get_youtube_transcript(video_url, lang="en"):
    ydl_opts = {
        'skip_download': True,
        'writesubtitles': True,
        'writeautomaticsub': True,
        'subtitlesformat': 'vtt',
        'subtitleslangs': [lang],
        'outtmpl': '%(id)s.%(ext)s',
        'quiet': True,
    }

    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(video_url, download=False)
        subs = info.get('subtitles', {}) or info.get('automatic_captions', {})
        if lang in subs:
            vtt_url = subs[lang][0]['url']
            vtt_text = requests.get(vtt_url).text
            return vtt_text
        else:
            return None

def vtt_to_text_preserve_lines(vtt_string):
    lines = vtt_string.splitlines()
    output = []
    current_line = []

    for line in lines:
        line = line.strip()
        if "-->" in line or line.isdigit():
            continue
        if line == "":
            if current_line:
                output.append(" ".join(current_line))
                current_line = []
        else:
            current_line.append(line)

    if current_line:
        output.append(" ".join(current_line))

    return "\n".join(output)

url = "https://www.youtube.com/watch?v=7owOKKgT0ak"
vtt_transcript = get_youtube_transcript(url, lang="en")

if vtt_transcript:
    text_transcript = vtt_to_text_preserve_lines(vtt_transcript)
    print(text_transcript)
else:
    print("Transcript not found for the specified language.")
```

```python
import json

file_path = '/content/drive/MyDrive/Colab Notebooks/transcript_2_fixed.json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
transcript_text = []

for event in data.get("events", []):
    for seg in event.get("segs", []):
        if "utf8" in seg:
            transcript_text.append(seg["utf8"])

final_transcript = "".join(transcript_text)

print(final_transcript)
```

```python
from rpunct import RestorePuncts

rpunct = RestorePuncts()
```

```python
restored_text = rpunct.punctuate(final_transcript)
print(restored_text)
```

```python
prompt = f'"Summarize this video: \ntext = "{restored_text}'
```
